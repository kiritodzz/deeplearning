## lets load the deep learning model

```python
mobile = tf.keras.applications.mobilenet_v2.MobileNetV2()  # deeplearning model -pre trained
```

## we have four phases in deep learning
## 1 Creating model 
## 2 Training a model
## 3 Testing a model
## 4 predict
## before predict we have to pre-processing of the image

```python
from tensorflow.keras.preprocessing import image
img = image.load_img(fileName,target_size = (224,224))
```

```python
plt.imshow(img)
```

## pre-processing of image
## Steps
## 1 resize image into array it is RGB therefore it will me (224,224,3) using tensorflow.keras.preprocessing import image
## 2 add 4th dimenision to resized image using numpy as for deeplearing we need four didimension
## 3 then pre-process using tf.keras.applications.mobilenet.preprocess_input(final image of 4th dimension)

```python
## step one
resized_img = image.img_to_array(img)
```

```python
# step two
final_image = np.expand_dims(resized_img, axis=0) #adding 4th dimension
```

```python
# step three
final_image = tf.keras.applications.mobilenet.preprocess_input(final_image)
```

```python
#size of final image
final_image.shape
```

```python
# size of resized image
resized_img.shape
```

## we haven't provided any label so to see our Result we need to use some util from tensorflow.keras.applications import imagenet_utils

```python
from tensorflow.keras.applications import imagenet_utils
predictions = mobile.predict(final_image)
results = imagenet_utils.decode_predictions(predictions)
print(results)
```
